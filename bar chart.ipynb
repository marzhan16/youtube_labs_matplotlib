```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
age_dev = [25, 26, 26, 27, 28, 29, 30]
dev_y = [300000, 320000, 380000, 410000, 450000, 500000, 520000]
x_indexes=np.arange(len(age_dev))
width = 0.25
plt.title('Average of salary')
plt.xlabel('The age of people')
plt.ylabel('The salary of people')
plt.bar(x_indexes-width, dev_y, color="m", width=width, label= "All developers")
py_dev_y = [280000, 340000, 370000, 430000, 480000, 510000, 550000]
plt.bar(x_indexes, py_dev_y, width=width, label='Python')
js_dev_y = [300000, 340000, 390000, 440000, 480000, 490000, 530000]
plt.bar(x_indexes+width, js_dev_y, width=width, label="JavaScript" )
plt.xticks(ticks=x_indexes, labels=age_dev)
plt.legend()
plt.show()
```
