```python
from matplotlib import pyplot as plt
import pandas as pd
plt.style.use('seaborn')

data = pd.read_csv('data1.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

#fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, sharex = True)
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, linestyle = '--', color = 'b', label = 'All developers')

ax2.plot(ages, py_salaries, color = 'yellow', label = 'Python')
ax2.plot(ages, js_salaries, color = 'g', label = 'JavaScript')

ax1.set_title('Median salary by age')
#ax1.set_xlabel('Age')
ax1.set_ylabel('Developers')
ax1.legend()

ax2.set_xlabel('Age')
ax2.set_ylabel('Developers')
ax2.legend()
plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')

```
